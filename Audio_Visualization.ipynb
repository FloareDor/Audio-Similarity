```python
import numpy as np
import pandas as pd
from glob import glob
import matplotlib.pyplot as plt

import librosa as lr
from scipy import spatial
import difflib
```

```python
data_dir = 'E:\FL\Data\Patches\Packs\Mura Masa Samples (Organised)\Music Loops'
audio_files = glob(data_dir + '/*.wav')
arr = []
```

```python
for j in audio_files:
    audio, sfreq = lr.load(j)
    print(j)
    time = np.arange(0, len(audio)) / sfreq
    fig, ax = plt.subplots()
    ax.plot(time, audio)
    ax.set(xlabel = 'Time (s)', ylabel = "Amplitude")
    #print(time, audio)
    arr.append(j)
    arr.append(time)
    arr.append(audio)
print(arr)
```

```python

path_to_sample = 'E:\FL\Data\Patches\Packs\Mura Masa Samples (Organised)\Music Loops\87_Ds_VividMusic_SP.wav'
n = arr.index(path_to_sample)
print(n)
st = arr[n+1].tolist()
sa = arr[n+2].tolist()
print(len(arr))
for i in range(0,len(arr),3):
    fig, ax = plt.subplots()
    ax.plot(arr[i+1], arr[i+2])
    ax.set(xlabel = 'Time (s)', ylabel = "Amplitude")
    plt.show()
    tt = arr[i+1]
    ta = arr[i+2]
    rt = difflib.SequenceMatcher(None, st, tt)
    ra = difflib.SequenceMatcher(None, sa, ta)
    score = (rt.ratio() + ra.ratio())/2
    print(f"Score: {score}")
    if i == len(arr) - 1:
        break
    
    






#sm=difflib.SequenceMatcher(None,s1,s2)
#print(sm.ratio())
#print(dataSetI)

#result = 1 - spatial.distance.cosine(dataSetI, dataSetII)
#print(result)
```
